```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1000

x = np.random.randn(n)

width = 0.5
histmin = np.floor(min(x))
histmax = np.ceil(max(x))+width

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=100,alpha=0.5,edgecolor="black")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.savefig("prompt6-histo.pdf",bbox_inches="tight",facecolor="white")
```
